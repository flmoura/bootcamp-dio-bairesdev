```python
from PIL import Image
```

```python
def rgb_to_grayscale(image_path):
    """
    Converte uma imagem RGB para escala de cinza.

    Args:
        image_path (str): O caminho para a imagem que será convertida.

    Returns:
        Image: A imagem convertida para escala de cinza.

    """
    image = Image.open(image_path)
    width, height = image.size

    for x in range(width):
        for y in range(height):
            r, g, b = image.getpixel((x, y))
            gray = int((r * 0.299) + (g * 0.587) + (b * 0.114))
            image.putpixel((x, y), (gray, gray, gray))

    return image
```

```python
def grayscale_to_black_and_white(image_path, threshold = 127):
    """
    Converte uma imagem em escala de cinza para preto e branco com base em um limiar.

    Args:
        image_path (str): O caminho para a imagem em escala de cinza.
        threshold (int, opcional): O valor do limiar para a conversão. Pixels com valor menor que o limiar serão convertidos para preto (0, 0, 0), 
                                    e pixels com valor maior ou igual ao limiar serão convertidos para branco (255, 255, 255). O valor padrão é 127.

    Returns:
        PIL.Image.Image: A imagem convertida em preto e branco.
    """
    image = rgb_to_grayscale(image_path)
    width, height = image.size

    for x in range(width):
        for y in range(height):
            color = image.getpixel((x, y))[0]
            if color < threshold:
                image.putpixel((x, y), (0, 0, 0))
            else:
                image.putpixel((x, y), (255, 255, 255))

    return image
```
